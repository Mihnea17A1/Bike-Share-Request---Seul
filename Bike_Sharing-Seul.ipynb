# Train/valid/test dataset

```python
train, val, test = np.split(df.sample(frac=1), [int(0.6*len(df)), int(0.8*len(df))])
```

```python
def get_xy(dataframe, y_label, x_labels=None):
    # Creez o copie profunda a dataframe-ului pentru a evita modificarea datelor originale
    dataframe = copy.deepcopy(dataframe)

    # Daca nu sunt furnizate etichete specifice pentru X (caracteristicile),
    # selectez toate coloanele din dataframe cu exceptia coloanei y_label
    if x_labels is None:
        X = dataframe[[c for c in dataframe.columns if c != y_label]].values
    else:
        # Daca este furnizata doar o singura eticheta pentru X, o selectez
        # si redimensionez rezultatul pentru a avea forma corecta
        if len(x_labels) == 1:
            X = dataframe[x_labels[0]].values.reshape(-1, 1)
        # Daca sunt furnizate mai multe etichete pentru X, selectez acele coloane
        else:
            X = dataframe[x_labels].values

    # Selectez coloana y_label ca variabila dependenta (tinta) si o redimensionez
    y = dataframe[y_label].values.reshape(-1, 1)

    # Combin matricea caracteristicilor (X) si vectorul tinta (y) intr-o singura matrice
    data = np.hstack((X, y))

    # Returneaz matricea nou formata --> matricea caracteristicilor si vectorul tinta
    return data, X, y
```

```python
_, X_train_temp, y_train_temp = get_xy(train, "bike_count", x_labels=["temp"])
_, X_val_temp, y_val_temp = get_xy(val, "bike_count", x_labels=["temp"])
_, X_test_temp, y_test_temp = get_xy(test, "bike_count", x_labels=["temp"])
```

```python
temp_reg = LinearRegression()
temp_reg.fit(X_train_temp, y_train_temp)
```

```python
# Evaluez performanta modelului de regresie liniară calculand
# coeficientul de determinare

temp_reg.score(X_test_temp, y_test_temp)
```

```python
plt.scatter(X_train_temp, y_train_temp, label="Data", color="blue")
x = tf.linspace(-20, 40, 100)
plt.plot(x, temp_reg.predict(np.array(x).reshape(-1, 1)), label="Fit", color="red", linewidth=1.5)
plt.legend()
plt.title("Bikes vs Temp")
plt.ylabel("Number of bikes")
plt.xlabel("Temp")
plt.show()
```
